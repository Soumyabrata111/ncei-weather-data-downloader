```python
# Change the url to iterate the year from 2010 to 2023 and store the URLs in a list
urls = []
for i in range(2010, 2024):
    urls.append('https://www.ncei.noaa.gov/data/global-hourly/access/' + str(i) + '/42809099999.csv')
print(urls)
```

```python
# Download the files from the URLs, store them in a list of dataframes, then concatenate
import pandas as pd
dfs = [pd.read_csv(url, skiprows=5, header=None) for url in urls]
df = pd.concat(dfs)
df.head()
```

```python
# Download the files from the URLs to separate dataframes
import pandas as pd
dfs = {url: pd.read_csv(url) for url in urls}
```

```python
# Merge the dataframes vertically
merged_df = pd.concat([dfs[urls[0]]] + [dfs[url].iloc[1:] for url in urls[1:]])
merged_df.head()
merged_df.to_csv('NETAJI SUBHASH CHANDRA BOSE INTERNATIONAL merged.csv')
```
